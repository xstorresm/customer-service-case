## Transform Categorical Data

```python
# Identify categorical features
categorical_columns = df.select_dtypes('object').columns
#print("\nCategorical features:")
categorical_columns
```

```python
# Create dummy variables for each categorical feature
for column in categorical_columns:
    # Create dummy variables with a prefix
    dummy_columns = pd.get_dummies(df[column], prefix=column, drop_first=True)
    # Convert True/False to 1/0
    dummy_columns = dummy_columns.astype(int)
    # Add the dummy columns to the original DataFrame
    df = pd.concat([df, dummy_columns], axis=1)
    # Drop the original categorical column from the DataFrame
    df.drop(column, axis=1, inplace=True)

# Display the modified DataFrame with dummy variables
#print("\nDataFrame with dummy variables:")
df.head()
```

# Applying Classification Method: Naive Bayes

```python
from sklearn.model_selection import train_test_split

features = df.drop(['satisfaction_satisfied'], axis=1)
response = df['satisfaction_satisfied']
features.shape
```

### Features Selection with LogisticRegression

```python
# Create pipeline that applys selection clasification and Standard Scaler 
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import StandardScaler

pipe_lr= Pipeline([
    ('logistic_regression', SelectFromModel(estimator = LogisticRegression(max_iter=5000))),
    ('Standard_scaler', StandardScaler())
])
 
pipe_lr.fit(features, response)
```

```python
# saving the  selected feautures with transformations 
X_1= pipe_lr.transform(features)
X_1.shape
```

```python
#Get the features selected 
features_selected_lr = pipe_lr['logistic_regression'].get_support(indices=True)

selected_feature_names_lr = features.columns[features_selected_lr]
#Create a new dataframe with features selected
df_features_selected_lr = pd.DataFrame(X_1, columns=selected_feature_names_lr)
#df_features_selected_lr.head()
df_features_selected_lr.columns
```

## Classifier Naive Bayes Model

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X_1, response, test_size=0.20, random_state=42)
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.naive_bayes import GaussianNB

#List of classifier
classifiers= [('naive_bayes', GaussianNB())]

clf_list=[]
list_accuracy=[]
kf = KFold(n_splits=5, shuffle=True, random_state=42)
best_classifier=0

#Pipleline  and evaluation of each classifier 
for n, clf in  classifiers:
    pipe_clf= Pipeline([(n,clf) ])
    score = cross_val_score(pipe_clf, X_train, y_train, cv=kf, scoring='accuracy')
    mean_score = np.mean(score)
    clf_list.append(n)
    list_accuracy.append((n, mean_score))
    
    if best_classifier < mean_score:
        best_clf_mean=mean_score
        best_clf=(n, clf)

X_2 = best_clf[1]
results = np.array(list_accuracy)
results_df = pd.DataFrame(results, columns=['Name', 'MeanScore'])
results_df
```

```python
X_2.fit(X_train, y_train)
```

## Confusion Matrix 

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix

y_pred = X_2.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
df_cm = pd.DataFrame(conf_matrix, columns=np.unique(y_test), index = np.unique(y_test))
df_cm.index.name = 'Actual'
df_cm.columns.name = 'Predicted'
plt.figure(figsize = (4,4))
sns.set(font_scale=1.2)#for label size
sns.heatmap(df_cm, annot=True,annot_kws={"size": 12}, cbar=False,
          vmax=5000, square=True, fmt="d", cmap="Reds")
```

## ROC Curve

```python
from sklearn.metrics import roc_auc_score 
from sklearn.metrics import roc_curve #plot the curve 

rf_roc_auc = roc_auc_score(y_test, X_2.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, X_2.predict_proba(X_test)[:,1]) 

plt.figure()
plt.plot(fpr, tpr, label='Naive Bayes Classifer (area = %0.2f)' % rf_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

# Clustering Method: KMeans

### 1.- Age and Flight Distance
In this first case, the variables Age and Flight Distance are selected for the clustering method with KMeans

```python
sns.scatterplot(data=df, x='Age', y='Flight Distance');
```

Standard scaler is applied to the variables to reduce the dominance effect, then the sum of the squares of the distances between each data point is calculated to create the Elbow curve.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

age = df["Age"]
flight_distance = df["Flight Distance"]

scaler = StandardScaler()
age_scaled = scaler.fit_transform(age.values.reshape(-1, 1))
flight_distance_scaled = scaler.fit_transform(flight_distance.values.reshape(-1, 1))

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k,n_init=10, random_state=42)
    kmeans.fit(np.column_stack((age_scaled, flight_distance_scaled)))
    inertia.append(kmeans.inertia_)

```

The Elbow curve is plotted to identify the optimal number of clusters.

```python
plt.plot(range(1, 11), inertia, "o-")
plt.xlabel("Number of clusters (K)")
plt.ylabel("Inertia")
plt.show()

```

According to the previous graph, it is identified that the optimal number of clusters is 3, so it is applied in the following graph.

```python
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
kmeans.fit(np.column_stack((age_scaled, flight_distance_scaled)))

# AsignaciÃ³n de las etiquetas de los clusters a los datos
df["Cluster"] = kmeans.labels_

# VisualizaciÃ³n de los resultados
plt.scatter(age_scaled, flight_distance_scaled, c=kmeans.labels_, s=50, alpha=0.5)
plt.xlabel("Age (standardized)")
plt.ylabel("Flight distance (standardized)")
plt.show()

```

In the next step, the centroids of each cluster are calculated and visually identified with an

```python

plt.scatter(age_scaled, flight_distance_scaled, c=kmeans.labels_, s=50, alpha=0.5)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c="red", s=100, marker="x")
plt.xlabel("Age (standardized)")
plt.ylabel("Flight distance (standardized)")
plt.show()
```
